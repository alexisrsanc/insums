```python
def asignar_categoria_uso(texto):
    """
    Asigna una categoría de uso según las palabras clave encontradas en el texto.

    Categorías:
    1: Infusión y Catéteres
    2: Guantes
    3: Jeringas y Agujas
    4: Sondas
    5: Tubos y Máscaras
    6: Apósitos y Vendas
    7: Equipos de Monitoreo y Accesorios
    8: Material de Ostomía
    9: Ropa y Protección Desechable
    10: Papel Médico
    11: Instrumental Quirúrgico
    12: Sistemas de Derivación y Válvulas
    13: Llaves y Conectores
    14: Material de Curación General
    15: Productos de Protección y Barrera
    16: Tubos de Aspiración y Succión
    17: Sistemas de Infusión y Administración
    18: Marcapasos y Dispositivos Electrónicos
    99: Otros (insumos no categorizados)
    """
    
    # Normalizar el texto
    texto = texto.lower().strip()

    # Diccionario de categorías y palabras clave
    categorias = {
        1: ['infusión', 'catéter', 'trocar', 'cánula', 'luer lock', 'venoso', 'intravenoso'],
        2: ['guante'],
        3: ['jeringa', 'aguja', 'insulina', 'aspiración de sangre'],
        4: ['sonda', 'alimentación enteral', 'nasogástrica', 'orogástrica', 'foley'],
        5: ['tubo', 'máscara', 'endotraqueal', 'oxígeno', 'humidificador', 'nebulizador'],
        6: ['apósito', 'venda', 'herida', 'quemadura', 'úlcera', 'hidrocoloide', 'hidrofibra', 'plata', 'curación'],
        7: ['monitorización', 'electrodo', 'sensor', 'oxímetro', 'ecg', 'electrocardiógrafo'],
        8: ['ostomía', 'bolsa drenable', 'placa colostomía', 'urostomía'],
        9: ['bata', 'cubrebocas', 'mascarilla', 'gorros', 'escudo facial', 'ropa desechable'],
        10: ['papel', 'ecográfico', 'electrocardiógrafo', 'monitor'],
        11: ['tijeras', 'pinza', 'porta bisturí', 'riñón acero', 'instrumental quirúrgico'],
        12: ['válvula hidrocefalia', 'válvula fonación', 'presión media', 'derivativa'],
        13: ['llave de tres vías', 'conector oxígeno', 'adaptador nebulización'],
        14: ['gasa estéril', 'curación general', 'compresa radiopaca'],
        15: ['pechera', 'cubrecalzado', 'escudo facial', 'barrera protección'],
        16: ['succión', 'aspiración', 'filtro humidificador'],
        17: ['equipo administración sangre', 'acelerador de infusión'],
        18: ['marcapasos', 'bicameral', 'unicameral'],
    }

    # Iterar sobre el diccionario para asignar la categoría correspondiente
    for cat_id, palabras_clave in categorias.items():
        if any(palabra in texto for palabra in palabras_clave):
            return cat_id

    return 99  # Si no se encuentra coincidencia, devuelve 99 (Otros)

```

```python
def asignar_funcion_especifica(texto):
    """
    Asigna un identificador de función específica según las palabras clave encontradas en el texto.
    
    Funciones específicas:
    1: Drenaje, Aspiración, Cirugía, Quirúrgico
    2: Protección, Cubrir, Aislamiento
    3: Diagnóstico, Monitoreo, Medición
    4: Curación, Tratamiento, Recuperación
    11: Instrumentos médicos y quirúrgicos
    12: Materiales desechables
    13: Equipos de oxígeno y ventilación
    14: Sistemas de drenaje y succión
    15: Papelería y textiles médicos
    16: Otros (sin clasificación)
    17: Productos de ostomía
    18: Accesorios de monitorización
    19: Accesorios de ventilación y traqueostomía
    20: Material quirúrgico desechable
    21: Adhesivos y apósitos avanzados
    22: Productos de limpieza médica
    23: Material ortopédico y de soporte
    24: Sistemas de succión y aspiración
    """

    # Asegurar que el texto es una cadena (manejo de errores si no lo es)
    if not isinstance(texto, str):
        texto = str(texto)

    # Normalizar el texto
    texto = texto.lower().strip()

    # Diccionario de funciones y palabras clave
    funciones = {
        1: ['drenaje', 'aspiración', 'cirugía', 'quirúrgico', 'quirúrgica'],
        2: ['protección', 'cubrir', 'aislamiento'],
        3: ['diagnóstico', 'monitoreo', 'medición', 'electrocardiógrafo'],
        4: ['curación', 'tratamiento', 'recuperación', 'herida', 'cicatrización'],
        11: ['pinza', 'tijera', 'porta bisturí', 'cánula', 'instrumento quirúrgico', 'trocar'],
        12: ['mascarilla', 'gorro', 'bata', 'desechable', 'cubrebocas'],
        13: ['oxígeno', 'ventilación', 'cánula guedel', 'mascarilla aerosol', 'conector oxígeno'],
        14: ['sonda', 'drenaje', 'aspiración', 'trocar', 'catéter'],
        15: ['papel', 'tela adhesiva', 'micropore', 'rollo', 'textil', 'algodón'],
        16: ['otros', 'sin categoría', 'sin clasificación'],  # Categoría por defecto
        17: ['ostomía', 'bolsa drenable', 'placa colostomía', 'urostomía'],
        18: ['monitorización', 'electrodo', 'sensor', 'oxímetro', 'monitor fetal', 'electrocardiógrafo'],
        19: ['cánula traqueostomía', 'mascarilla oxígeno', 'ventilación', 'traqueostomía'],
        20: ['cubre calzado', 'gorro', 'mascarilla desechable', 'bata quirúrgica'],
        21: ['adhesivo', 'apósito', 'hidrocoloide', 'hidrofibra', 'tela adhesiva'],
        22: ['cepillo limpieza', 'eliminador adhesivos', 'spray limpieza', 'desinfectante'],
        23: ['yeso', 'tejido tubular', 'material ortopédico', 'soporte'],
        24: ['succión', 'aspiración', 'tubo conector', 'sistemas succión'],
    }

    # Iterar sobre el diccionario para asignar la función específica
    for funcion, palabras_clave in funciones.items():
        if any(palabra in texto for palabra in palabras_clave):
            return funcion

    # Si no se encuentra coincidencia, asignar la categoría "Otros"
    return 16
```

```python
def asignar_esterilidad(texto):
    """
    Asigna un identificador de esterilidad según las palabras clave encontradas en el texto.

    Esterilidad Categorías:
    1: Esteril
    2: No esteril
    3: Desechable
    4: No desechable / Reutilizable
    5: Productos médicos avanzados
    6: Accesorios para monitorización y diagnóstico
    7: Oxigenoterapia y ventilación
    8: Productos para curación y tratamiento
    9: Productos para drenaje y succión
    10: Material quirúrgico y de procedimiento
    11: Material desechable general
    12: Otros
    """
    texto = texto.lower().strip()

    categorias = {
        1: ['esteril'],
        2: ['no esteril'],
        3: ['desechable'],
        4: ['reutilizable', 'inoxidable'],
        5: ['marcapasos', 'válvula', 'traqueostomía', 'set traqueostomía'],
        6: ['monitorización', 'diagnóstico', 'electrocardiógrafo', 'ecg', 'papel ecg', 'sensor'],
        7: ['oxígeno', 'ventilación', 'cánula guedel', 'tubo endotraqueal', 'máscara de oxígeno'],
        8: ['curación', 'tratamiento', 'apósito', 'venda', 'gel hidroactivo', 'hidrocoloide'],
        9: ['drenaje', 'aspiración', 'trocar', 'sonda foley', 'catéter'],
        10: ['quirúrgico', 'pinza', 'tijera', 'porta bisturí'],
        11: ['cubre calzado', 'gorro', 'mascarilla'],
        12: ['otros']
    }

    for categoria, palabras_clave in categorias.items():
        if any(palabra in texto for palabra in palabras_clave):
            return categoria

    return 12  # Si no se encuentra coincidencia, asignar "Otros"


# Aplicar la función al DataFrame
df['esterilidad_id'] = df['descripcion_completa'].apply(asignar_esterilidad)

# Verificar resultados
print("Distribución de categorías:")
print(df['esterilidad_id'].value_counts())
```

```python
def asignar_material(texto):
    """
    Asigna el nombre de la categoría de material según la primera palabra clave encontrada en el texto.

    Si aparecen múltiples palabras clave de distintas categorías, se devolverá el nombre
    correspondiente a la palabra clave cuya posición en el texto sea más temprana.
    """

    texto = texto.lower().strip()

    categorias = {
        'Guantes': ['guante'],
        'Jeringas y Agujas': ['jeringa', 'aguja'],
        'Catéteres y Vías': ['cánula', 'tubo endotraqueal', 'microlaríngeo'],
        'Sondas y Catéteres Urinarios': ['sonda', 'foley', 'pezzer', 'nelaton'],
        'Apósitos y Vendajes': ['apósito', 'gasa', 'compresa', 'curación'],
        'Instrumental Quirúrgico': ['pinza', 'tijera', 'porta bisturí', 'riñón quirúrgico', 'copela'],
        'Infusión y Succión': ['infusión', 'succión', 'aspiración', 'regulador de goteo'],
        'Oxigenoterapia': ['oxígeno', 'ventilación', 'máscara', 'humidificador'],
        'Monitoreo y Sensores': ['papel', 'ecg', 'sensor', 'transductor'],
        'Protección Personal': ['bata', 'cubre calzado', 'gorro', 'mascarilla'],
        'Material de Ostomía': ['ostomía', 'bolsa', 'placa', 'pasta'],
        'Vendajes Especiales': ['venda', 'yeso', 'tricotaje', 'tejido tubular'],
        'Limpieza y Adhesivos': ['limpieza', 'cepillo', 'adhesivo'],
        'Dispositivos Electrónicos': ['marcapasos', 'traqueostomía', 'válvula de fonación'],
    }

    for categoria, palabras_clave in categorias.items():
        if any(palabra in texto for palabra in palabras_clave):
            return categoria

    return 'Otros'  # Si no se encuentra coincidencia, asignar "Otros"

# Aplicar la función al DataFrame
df['material_id'] = df['descripcion_completa'].apply(asignar_material)
# Verificar resultados
print("Distribución de categorías:")
print(df['material_id'].value_counts())
```
