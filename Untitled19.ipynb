```python
import pandas as pd

# Paso 1: Construir la columna 'descripcion_completa'
df['descripcion_completa'] = (
    df['DENOMINACIÓN'].astype(str) + ' ' + df['DESCRIPCIÓN DEL PRODUCTO OFERTADO'].astype(str)
).str.lower()

# Paso 2: Aplicar funciones de clasificación
df['categoriaUso_id'] = df['descripcion_completa'].apply(asignar_categoria_uso)
df['material_id'] = df['descripcion_completa'].apply(asignar_material)
df['esterilidad_id'] = df['descripcion_completa'].apply(asignar_esterilidad)
df['tamano_id'] = df['descripcion_completa'].apply(asignar_tamano)
df['funcionEspecifica_id'] = df['descripcion_completa'].apply(asignar_funcion_especifica)

# Paso 3: Exportar Excel por cada clasificación con ID == 0
columnas_clasificacion = [
    'categoriaUso_id', 'material_id', 'esterilidad_id', 'tamano_id', 'funcionEspecifica_id'
]

for columna in columnas_clasificacion:
    df_filtrado = df[df[columna] == 0]
    if not df_filtrado.empty:
        # Columnas a incluir en el Excel
        columnas_exportar = [
            'DENOMINACIÓN',
            'DESCRIPCIÓN DEL PRODUCTO OFERTADO',
            'descripcion_completa',
            columna
        ]
        nombre_archivo = f"productos_{columna}_0.xlsx"
        df_filtrado[columnas_exportar].to_excel(nombre_archivo, index=False)
        print(f"Archivo generado: {nombre_archivo} ({len(df_filtrado)} registros)")
    else:
        print(f"No se encontraron registros con {columna} == 0")
```

```python
def asignar_categoria_uso(texto):
    """
    Asigna una categoría de uso según las palabras clave encontradas en el texto.
    Categorías:
    1: Infusión y Catéteres
    2: Guantes
    3: Jeringas y Agujas
    4: Sondas
    5: Tubos y Máscaras
    6: Apositos y Vendas
    7: Equipos de Monitoreo y Accesorios
    8: Material de Ostomía
    9: Ropa y Protección Desechable
    10: Papel Médico
    0: Sin clasificación
    """
    # Normalizar el texto
    texto = texto.lower().strip()
    
    # Diccionario de categorías y palabras clave
    categorias = {
        1: ['infusión', 'catéter', 'trocar', 'cánula', 'acelerador de infusión', 'luer lock', 'venoso', 'intravenoso', 'umbilical'],
        2: ['guante'],
        3: ['jeringa', 'aguja', 'insulina', 'aspiración de sangre'],
        4: ['sonda', 'alimentación enteral', 'nasogástrica', 'orogástrica', 'foley'],
        5: ['tubo', 'máscara', 'endotraqueal', 'oxígeno', 'humidificador', 'nebulizador'],
        6: ['apósito', 'venda', 'herida', 'quemadura', 'úlcera', 'hidrocoloide', 'hidrofibra', 'plata', 'curación'],
        7: ['monitorización', 'electrodo', 'sensor', 'oxímetro', 'ecg', 'papel ecg', 'electrocardiógrafo'],
        8: ['ostomía', 'bolsa drenable', 'placa colostomía', 'urostomía'],
        9: ['bata', 'cubrebocas', 'mascarilla', 'gorros', 'escudo facial', 'bata quirúrgica', 'ropa desechable'],
        10: ['papel', 'ecográfico', 'electrocardiógrafo', 'monitor'],
    }
    
    # Asignar categoría
    for categoria, palabras_clave in categorias.items():
        if any(palabra in texto for palabra in palabras_clave):
            return categoria
    
    # Sin clasificación
    return 0
```

```python
def asignar_funcion_tamano(texto):
    """
    Asigna una función específica según las palabras clave relacionadas con el tamaño encontradas en el texto.
    Funciones:
    1: Tamaño pequeño
    2: Tamaño mediano
    3: Tamaño grande
    4: Tamaño extra grande
    0: Sin clasificación
    """
    # Normalizar el texto
    texto = texto.lower().strip()
    
    # Diccionario de funciones y palabras clave
    funciones = {
        1: ['pequeño', 'small', 'xs', 'chico', 'mini'],
        2: ['mediano', 'm', 'medium'],
        3: ['grande', 'l', 'large', 'xl'],
        4: ['extra grande', 'xxl', 'xxxl', 'gigante', 'extra large']
    }
    
    # Asignar función específica
    for funcion, palabras_clave in funciones.items():
        if any(palabra in texto for palabra in palabras_clave):
            return funcion
    
    # Sin clasificación
    return 0
```
